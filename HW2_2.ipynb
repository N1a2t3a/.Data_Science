```python
import pandas as pd

table_info = pd.read_csv('/content/2017_jun_final (1).csv')
```

```python
columns_to_drop = table_info[table_info.columns.difference(['Язык.программирования'])].isna().any()
table_info = table_info.drop(columns_to_drop.index[columns_to_drop], axis=1)
table_info
```

```python
table_info.isnull().sum()
```

```python
table_info = table_info.dropna(axis=0)
table_info
```

```python
python_data = table_info[table_info["Язык.программирования"] == "Python"]
python_data
```

```python
size_of_python_data = python_data.shape
print("Розмір таблиці python_data:", size_of_python_data)
```

```python
grouped_data = table_info.groupby('Должность')
```

```python
aggregated_data = table_info.groupby('Должность')['Зарплата.в.месяц'].agg(['min', 'max']).reset_index()
```

```python
table_info['Зарплата.в.месяц'] = table_info['Зарплата.в.месяц'].astype(str).str.replace(r'\D+', '')

table_info['Зарплата.в.месяц'] = pd.to_numeric(table_info['Зарплата.в.месяц'], errors='coerce')

table_info['avg'] = table_info.mean(axis=1)
```

```python
avg_stats = table_info['avg'].describe()
print(avg_stats)
```
